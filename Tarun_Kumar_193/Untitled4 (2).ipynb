```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(42)
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.randn(100, 1)


X_b = np.c_[np.ones((100, 1)), x]


theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

print("Intercept (theta0):", theta_best[0][0])
print("Slope (theta1):", theta_best[1][0])


x_new = np.array([[0], [2]])
x_new_b = np.c_[np.ones((2, 1)), x_new]
y_predict = x_new_b.dot(theta_best)


plt.scatter(x, y)
plt.plot(x_new, y_predict)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression Curve")
plt.show()
```

```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target

print(df.head())
```

```python
import matplotlib.pyplot as plt

class_counts = df['target'].value_counts()

class_counts.plot(kind='bar')

plt.title("Count of Each Iris Class")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
import matplotlib.pyplot as plt

avg_values = df.groupby('target')['sepal length (cm)'].mean()

avg_values.plot(kind='bar')

plt.title("Average Sepal Length per Class")
plt.xlabel("Class")
plt.ylabel("Average Sepal Length")
plt.show()
```

```python
import matplotlib.pyplot as plt

df['sepal length (cm)'].plot(kind='hist', bins=10)

plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```
