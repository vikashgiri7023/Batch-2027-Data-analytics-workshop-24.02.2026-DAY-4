1.Drawing Bar Charts (draw bar charts using Pandas and matplotlib.) 2. Making Histograms (draw histograms using Pandas and matplotlib.)

```python
import pandas as pd
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()

# Convert to DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add target column
df['target'] = iris.target

# Display first 5 rows
print(df.head())
```

```python
import matplotlib.pyplot as plt

# Count target values
class_counts = df['target'].value_counts()

# Plot bar chart
class_counts.plot(kind='bar')

plt.title("Count of Each Iris Class")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Average sepal length per class
avg_values = df.groupby('target')['sepal length (cm)'].mean()

# Plot bar chart
avg_values.plot(kind='bar')

plt.title("Average Sepal Length per Class")
plt.xlabel("Class")
plt.ylabel("Average Sepal Length")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Histogram of sepal length
df['sepal length (cm)'].plot(kind='hist', bins=10)

plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Histogram for all features
df.hist(figsize=(8, 6))

plt.show()
```

Drawing Box Plots (draw box plots using Pandas and matplotlib.)
2. Drawing Some Other Kinds of Plots with Matplotlib (Learn about some of the other charts
supported by matplotlib.)


```python
import pandas as pd
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()

# Convert to DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add target column
df['target'] = iris.target

# Display first 5 rows
print(df.head())
```

```python
import matplotlib.pyplot as plt

# Box plot for sepal length
df.boxplot(column='sepal length (cm)')

plt.title("Box Plot of Sepal Length")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Box Plot for All Features
df.iloc[:, 0:4].boxplot(figsize=(8,6))
plt.title("Box Plot of All Features")
plt.show()







# Pie Chart
df['target'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Class Distribution")
plt.ylabel("")
plt.show()

# Area Plot
df.iloc[:, 0:4].plot.area(alpha=0.4)
plt.title("Area Plot")
plt.show()
```

```python
# Class-wise Box Plot
df.boxplot(column='sepal length (cm)', by='target')
plt.title("Sepal Length by Class")
plt.suptitle("")
plt.show()
```

```python
# Line Plot
plt.plot(df['sepal length (cm)'])
plt.title("Line Plot of Sepal Length")
plt.xlabel("Index")
plt.ylabel("Sepal Length")
plt.show()
```

```python
# Scatter Plot
plt.scatter(df['sepal length (cm)'], df['petal length (cm)'])
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.title("Scatter Plot")
plt.show()
```
